```python
df_converted1 = df_converted.loc[df_converted['tier'] == 1]
df_converted2 = df_converted.loc[df_converted['tier'] == 2]
df_converted3 = df_converted.loc[df_converted['tier'] == 3]
```

```python
# look at tier 1
cc1=df_converted1.groupby('LastValue').count()['convert_TF']
df_cc1 = cc1.rename_axis("LastValue").reset_index(name = "convert_TF")
df_cc1['Spending'] = [0,1000,1000,0,1000,1000,1000] 
df_cc1['CAC'] = df_cc1['Spending']/df_cc1['convert_TF']
df_cc1
#cc1 stands for converted customers for tier 1
```

```python
# look at tier 2
cc2=df_converted2.groupby('LastValue').count()['convert_TF']
df_cc2 = cc2.rename_axis("LastValue").reset_index(name = "convert_TF")
df_cc2['Spending'] = [0,2000,2000,0,2000,2000,2000] 
df_cc2['CAC'] = df_cc2['Spending']/df_cc2['convert_TF']
df_cc2
```

```python
# look at tier 3
cc3=df_converted3.groupby('LastValue').count()['convert_TF']
df_cc3 = cc3.rename_axis("LastValue").reset_index(name = "convert_TF")
df_cc3['Spending'] = [0,3000,3000,0,3000,3000,3000] 
df_cc3['CAC'] = df_cc3['Spending']/df_cc3['convert_TF']
df_cc3
```

# 1 c)

```python
marg_cac1=[]
marg_cac2=[]
for i in range(7):
    a = (df_cc2.loc[i,'Spending']-df_cc1.loc[i,'Spending'])/(df_cc2.loc[i,'convert_TF']-df_cc1.loc[i,'convert_TF'])
    b = (df_cc3.loc[i,'Spending']-df_cc2.loc[i,'Spending'])/(df_cc3.loc[i,'convert_TF']-df_cc2.loc[i,'convert_TF'])
    marg_cac1.append(a)
    marg_cac2.append(b)
```
